```python
# Dropping irrelevant columns for analysis
data.drop(columns=['brokered_by','street'], inplace=True, errors='ignore')
```

```python
data.head()
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# Filling missing values with median for numerical columns
num_cols = ['price','bed','bath','acre_lot','house_size']
for col in num_cols:
    if col in data.columns:
        data[col] = data[col].fillna(data[col].median())
```

```python
# Fill missing values with mode for categorical columns
cat_cols = ['status','city','state']
for col in cat_cols:
    if col in data.columns:
        data[col] = data[col].fillna(data[col].mode())
```

```python
data.isnull().sum()
```

```python
# Dropping rows with missing values in critical columns
data.dropna(subset=['zip_code','prev_sold_date'], inplace=True)
```

```python
data.isnull().sum() 
```

```python
# Changing data types of certain columns
data['zip_code'] = data['zip_code'].astype('Int64', errors='ignore')
data['price'] = data['price'].astype('Int64', errors='ignore')
data['prev_sold_date'] = pd.to_datetime(data['prev_sold_date'], errors='coerce')
```

```python
data.info()
```

```python
# Checking for duplicates
data.duplicated().sum()
```

```python
# Dropping duplicates
data.drop_duplicates(inplace=True)
```

```python
# resetting index after dropping rows
data.reset_index(drop=True, inplace=True)
data.head()
```

<div style="background-color: #00e7c8ff; padding: 25px; border-radius: 10px;font-size:20px;font-weight:10px">
  <h1 style="color: #fff;" align="center"><strong> Analyzing Outliers</strong></h1>
</div>

```python
# Find Outliers using IQR method
def analyze_outliers(series):
    q1 = series.quantile(0.25)  # First quartile
    q2 =series.quantile(0.50)  # Second quartile (median)
    q3 = series.quantile(0.75)  # Third quartile
    iqr = q3-q1 # Interquartile range
    lower_bound = q1 - 1.5 * iqr  # Lower bound for outliers
    upper_bound = q3 + 1.5 * iqr  # Upper bound for outliers
    outlier = series[(series < lower_bound) | (series > upper_bound)]  # Identify outliers
    summary={
        "25 th percentile": q1,
        "50 th percentile": q2,
        "_75 th percentile": q3,
        "IQR": iqr,
        "Lower bound": lower_bound,
        "Upper bound": upper_bound,
        "Outlier_Count": outlier.size,
        "Outliers List": np.sort(outlier.unique())
    }
    return summary
```

```python
# Outlier from price column
analyze_outliers(data['price'])
```

```python
# analysing upper bound for price column
analyze_outliers(data['price'])['Upper bound']
```

```python
# Remove Outliers from price column
removed_outliers = data[data['price'] <= analyze_outliers(data['price'])['Upper bound']]
```

```python
removed_outliers.describe() # Get descriptive statistics of the dataframe after removing outliers
```

```python
# Top 10 states by number of listings
top_cities = data['city'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_cities.index, y=top_cities.values,hue=top_cities.index, color='steelblue', palette="Set2")
plt.title("Top 10 Cities by Number of Listings")
plt.ylabel("Number of Listings")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.show()
```

```python
# Price distribution 
plt.figure(figsize=(10,6))
sns.histplot(removed_outliers['price'], bins=30, kde=True, color='#00A699')
plt.title('Price Distribution after Removing Outliers')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
top_states = data['state'].value_counts().head(10)
sns.barplot(x=top_states.index, y=top_states.values,hue=top_states.index, color='steelblue', palette="viridis")
plt.title("Top 10 States by Number of Listings")
plt.ylabel("Number of Listings")
plt.xlabel("State")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 8 states by number of listings (share %)
top_states = data['state'].value_counts().head(8)

plt.figure(figsize=(8,8))
plt.pie(top_states.values, labels=top_states.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Top 8 States by Listings (Share %)")
plt.show()
```

```python
# Median house price by state (top 10)
state_prices = data.groupby('state')['price'].median().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.lineplot(x=state_prices.index, y=state_prices.values, marker="o", color="steelblue", markersize=10)
plt.title("Median House Price by State (Top 10)")
plt.ylabel("Median Price")
plt.xlabel("State")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 states by median price per sqft
data['price_per_sqft'] = data['price'] / data['house_size']
plt.figure(figsize=(12,6))
state_pps = data.groupby('state')['price_per_sqft'].median().sort_values(ascending=False).head(10)
sns.barplot(x=state_pps.values, y=state_pps.index,hue=state_pps.index, color='steelblue', palette="coolwarm",orient="h")
plt.title("Top 10 States by Median Price per Sqft")
plt.ylabel("State")
plt.xlabel("Median Price per Sqft")
plt.xticks(rotation=45)
plt.show() 
```

```python
def price_category(price):
    if price < 200000:
        return "Affordable"
    elif price < 700000:
        return "Mid-range"
    else:
        return "Luxury"

data['price_category'] = data['price'].apply(price_category)

plt.figure(figsize=(8,6))
sns.countplot(x='price_category', data=data, hue='price_category', palette="Set2")
plt.title("Distribution of Houses by Price Category")
plt.ylabel("Number of Houses")
plt.xlabel("Category")
plt.show()
```

```python
# Number of listings vs Median Price (Top 10 States)
state_counts = data['state'].value_counts().head(10)
state_prices = data.groupby('state')['price'].median().loc[state_counts.index] 

plt.figure(figsize=(12,6))
plt.plot(state_counts.index, state_counts.values, marker='o', label='Number of Listings', color='brown')
plt.plot(state_prices.index, state_prices.values, marker='o', label='Median Price', color='teal')
plt.title("Number of Listings vs Median Price (Top 10 States)")
plt.xlabel("State")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
